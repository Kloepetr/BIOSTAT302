```python
import numpy as np
from scipy.stats import norm

scores = np.array([70, 82, 63, 91, 57, 85, 65, 73, 84, 93])
mean = 76  
sigma = 12  

#1
z_scores = (scores - mean) / sigma
print("\nStandardized Scores (Z-scores):")
print(z_scores)

#2
z_cutoff = norm.ppf(0.85)  # Z-score for the 85th percentile
cutoff_score = mean + z_cutoff * sigma

print("\nCutoff Score for an A:", round(cutoff_score, 2))

#3
students_with_A = scores[scores > cutoff_score]
print("Students who earned an A ".format(cutoff_score))
print(students_with_A)
```
