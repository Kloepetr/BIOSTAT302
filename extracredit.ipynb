```python
#Create a scatter plot comparing Ozone to Wind (0.5 point) and Ozone to temperature (0.5 point).
#import extensions
import sys
import scipy
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# read csv
airquality = pd.read_csv("airquality.csv")

# Ozone vs Wind
plt.scatter(airquality['Wind'], airquality['Ozone'])
plt.title('Ozone vs Wind')
plt.xlabel('Wind')
plt.ylabel('Ozone')
plt.show()

# Ozone vs Temperature
plt.scatter(airquality['Temp'], airquality['Ozone'])
plt.title('Ozone vs Temperature')
plt.xlabel('Temperature')
plt.ylabel('Ozone')
plt.show()
```

```python
'''Does ozone change by month? Perform an ANOVA analysis to see if Ozone changes by month (please perform all of the steps as outlined in lecture). 
Remember to include: hypotheses, F statistic, p-value, conclusions, and follow-up pairwise tests (if required). '''
#imports
import seaborn as sns
from scipy.stats import f_oneway


airquality_clean = airquality.dropna(subset=['Ozone', 'Month'])
groups = [group['Ozone'].values for name, group in airquality_clean.groupby('Month')]

# ANOVA
f_stat, p_val = f_oneway(*groups)
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.6f}")


#boxplot for pairwise test
plt.figure(figsize=(10, 6))
sns.boxplot(x='Month', y='Ozone', data=airquality_clean, palette='Set3')
plt.title('Boxplot of Ozone Levels by Month')
plt.xlabel('Month')
plt.ylabel('Ozone Level')
plt.show()
```
