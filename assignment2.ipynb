```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('CLABSI data.csv')

# Calculate the rate of infection
df['Rate of Infection'] = 1000 * df['Number of Infections'] / df['Patient Line Days']

# Print the institution name along with the rate of infection
for index, row in df.iterrows():
    print(f"{row['ICU Type']} - Rate of Infection: {row['Rate of Infection']:.2f}")
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('CLABSI data.csv')

# Calculate the rate of infection
df['Device Utilization Rate'] = 100 * df['Patient Line Days'] / df['Patient Days']

# Print the institution name along with the rate of infection
for index, row in df.iterrows():
    print(f"{row['ICU Type']} - Device Utilization Rate: {row['Device Utilization Rate']:.2f}")
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("LOSdata.csv") 
mean_value = df["LOS"].mean() 
std_deviation = np.std(df["LOS"])
tenth_percentile = np.percentile(df["LOS"], 10)
twenty_fifth_percentile = np.percentile(df["LOS"], 25)
fifthith_percentile = np.percentile(df["LOS"], 50)
seventy_fifth_percentile = np.percentile(df["LOS"], 75)
nighty_percentile = np.percentile(df["LOS"], 90)
nighty_fifth_percentile_percentile = np.percentile(df["LOS"], 95)
nighty_nine_percentile = np.percentile(df["LOS"], 99)
print("Mean value:", int(mean_value)) 
print("standard deviation", std_deviation)
print("10th percentile:", tenth_percentile)
print("25th percentile", twenty_fifth_percentile)
print("50th percentile:", fifthith_percentile)
print("75th percentile:", seventy_fifth_percentile)
print("90th percentile:", nighty_percentile)
print("95th percentile:", nighty_fifth_percentile_percentile)
print("99th percentile:", nighty_nine_percentile)
```
