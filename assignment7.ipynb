```python
# 3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from io import StringIO
df = pd.read_csv("pulse.csv")

# Optional: make 'exercise' column categorical for nicer labels
df['exercise'] = df['exercise'].astype('category')

# Create side-by-side boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='exercise', y='pulse', data=df)
plt.title('Pulse Rates by Exercise Group')
plt.xlabel('Exercise Group')
plt.ylabel('Pulse')
plt.grid(True)
plt.show()
```

```python
#4 and 5
import pandas as pd

# Load your data
df = pd.read_csv("pulse.csv")

# Group by exercise level and calculate summary stats
summary_stats = df.groupby('exercise')['pulse'].agg(
    sample_size='count',
    mean='mean',
    std_dev='std'
).reset_index()
print(summary_stats)


df['exercise'] = df['exercise'].astype('category')
plt.figure(figsize=(8, 6))
sns.pointplot(x='exercise', y='pulse', data=df, ci='sd', capsize=0.2)

plt.title('Mean Pulse by Exercise Group')
plt.xlabel('Exercise Group')
plt.ylabel('Mean Pulse')
plt.show()
```

```python
#6 perform levene test 
levene_test = stats.levene(df[df['exercise'] == 1]['pulse'], 
                           df[df['exercise'] == 2]['pulse'],
                           df[df['exercise'] == 3]['pulse'])

# Print the result of Levene's Test
print(f"Levene's Test p-value: {levene_test.pvalue}")
```

```python
#8
anova_result = stats.f_oneway(df[df['exercise'] == 1]['pulse'],
                              df[df['exercise'] == 2]['pulse'],
                              df[df['exercise'] == 3]['pulse'])

print(f"ANOVA F-statistic: {anova_result.statistic}")
print(f"ANOVA p-value: {anova_result.pvalue}")

# Interpretation
if anova_result.pvalue < 0.05:
    print("There is a significant difference in pulse rates across exercise groups.")
else:
    print("There is no significant difference in pulse rates across exercise groups.")
```
