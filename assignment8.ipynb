```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Load data
df = pd.read_csv('tobacco.csv')

x = df['tobacco']
y = df['alcohol']

# Scatter plot
plt.scatter(x, y, color='blue', label='Data Points')

# Least Squares Regression Line
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='red', label=f'LSRL: y = {m:.2f}x + {b:.2f}')

plt.xlabel('Tobacco')
plt.ylabel('Alcohol')
plt.title('Scatter Plot with LSRL')
plt.legend()
plt.grid(True)
plt.show()

# Reshape x for sklearn (must be 2D)
x_reshaped = x.values.reshape(-1, 1)

# Linear regression model
model = LinearRegression()
model.fit(x_reshaped, y)

# Get R^2 value
r_squared = model.score(x_reshaped, y)
print(f"R-squared: {r_squared:.4f}")
print(f"Percentage of variation explained: {r_squared * 100:.2f}%")

```

```
import statsmodels.api as sm

# Prepare x and y
x = df['tobacco']
y = df['alcohol']

# Add constant for intercept
x_with_const = sm.add_constant(x)

# Fit OLS model
model = sm.OLS(y, x_with_const).fit()

# Print the summary
print(model.summary())

# Get confidence intervals
conf_int = model.conf_int(alpha=0.05)  # 95% CI
slope_ci = conf_int.loc['tobacco']
print(f"95% CI for the slope: [{slope_ci[0]:.4f}, {slope_ci[1]:.4f}]")
```
